# ðŸŒ¸: Plot Customization(Sample Diamonds Data)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Diamonds dataset (raw CSV link from seaborn-data GitHub)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"

# Load dataset
df = pd.read_csv(url)

# Show first 5 rows
print(df.head())
print(df.shape)
```

```python
print(f"Orginal Shape(Before):",{df.shape[0]})
print(f"Orginal Shape(Before):",{df.shape[1]})
print(f"{50*'-'}")
df = df.sample(5000)
print(f"New Shape(After):", {df.shape[0]})
print(f"New Shape(After):",{df.shape[1]})
```

## ðŸŒ¸: Setting Title

```python
sns.scatterplot(x='carat', y='price', data=df)
plt.title("Diamonds Price vs Carat")
plt.show() 
```

## ðŸŒ¸: Settings Xlabels

```python
sns.scatterplot(x='carat', y='price', data=df)
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show() 
```

## ðŸŒ¸: Customizing The Legends

```python
sns.scatterplot(x='carat', y='price', data=df, hue='cut')
plt.title("Diamonds Price vs Carat")
plt.xlabel("Carat Weights") 
plt.ylabel("Price Dollar")
plt.legend(title='Cut Type', loc='upper right',bbox_to_anchor=(1,1))
plt.show()
```

# ðŸŒ¸: Customizing Ticks    

```python
# Set Seaborn style
sns.set_theme(style="whitegrid")

# Plot histogram
plt.figure(figsize=(12, 6))
sns.histplot(
    df['price'],
    bins=40,                 # more bins for better distribution
    edgecolor='black',
    color='skyblue',
    kde=True                  # optional: show density curve
)

# Titles and labels
plt.title("Distribution of Diamond Prices", fontsize=16, weight='bold')
plt.xlabel("Price ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Customize ticks
plt.xticks(np.arange(0, 20001, 2000), rotation=45)
plt.yticks(np.arange(0, 2200, 200))

# Add grid lines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## ðŸŒ¸: Add Griding

```python

# -----------------------------
# Professional Seaborn Styling
# -----------------------------
sns.set_theme(
    style="whitegrid",   # "whitegrid", "darkgrid", "ticks", "white"
    context="talk",      # scales fonts for presentations
    palette="Set2"       # color palette for 'cut'
)

# Figure size
plt.figure(figsize=(12, 7))

# Scatter plot
sns.scatterplot(
    x='carat',
    y='price',
    data=df,
    hue='cut',
    style='cut',        # different markers per cut
    s=60,               # marker size
    alpha=0.6,          # transparency for overlapping points
    edgecolor='k'       # black edge for better visibility
)

# Titles and labels
plt.title("Diamond Prices vs Carat Weights", fontsize=18, weight='bold', pad=15)
plt.xlabel("Carat Weight", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)

# Legend customization
plt.legend(
    title='Cut Type',
    title_fontsize=12,
    fontsize=10,
    loc='upper left',
    frameon=True,
    facecolor='white',
    edgecolor='black'
)

# Grid lines
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Tight layout for spacing
plt.tight_layout()

plt.show()
```

## ðŸŒ¸: Customizing Spins

```python
# -----------------------------
# Seaborn Theme
# -----------------------------
sns.set_theme(
    style="darkgrid",   # options: whitegrid, darkgrid, ticks, white
    context="talk",      # scales fonts for presentations
    palette="Set2"       # color palette for categorical features
)

# Figure size
plt.figure(figsize=(12, 7))

# Box plot
ax = sns.boxplot(
    x='cut',
    y='price',
    data=df,
    palette='Set2',
    showfliers=True,      # show outliers
    linewidth=1.5,        # thicker box lines
    notch=True,           # optional: add notch to show confidence interval
    saturation=0.8        # makes box colors slightly transparent
)

# Remove top and right spines for cleaner look
sns.despine(top=True, right=True)

# Titles and labels
plt.title("Diamond Prices by Cut Type", fontsize=18, weight='bold', pad=20)
plt.xlabel("Cut Type", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)

# Grid customization
plt.grid(
    axis='y',            # only horizontal grid
    linestyle='--',
    linewidth=0.8,
    alpha=0.6,
    color='Black'         # works well on darkgrid
)

# Optional: Legend for boxplot (if using hue)
# For simple boxplot by 'cut', legend is not required
# If using hue, uncomment below:
# plt.legend(title='Cut Type', loc='upper right', edgecolor='black')

# Tight layout for spacing
plt.tight_layout()

plt.show()
```

## ðŸŒ¸: Set Figure Size

```python
# Figure size
plt.figure(figsize=(14, 8))  # improved larger figure for clarity

# Bar plot
sns.barplot(
    x="cut",
    y="price",
    data=df,
    ci=None,             # remove confidence interval for clean look
    edgecolor="black",   # border around bars
    alpha=0.85           # slight transparency
)

# Titles and labels
plt.title("Average Diamond Prices by Cut Type", fontsize=18, weight='bold', pad=20)
plt.xlabel("Cut Type", fontsize=14)
plt.ylabel("Average Price ($)", fontsize=14)

# Grid customization
plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.7
)

# Tight layout for spacing
plt.tight_layout()

plt.show()
```

## ðŸŒ¸: SubPlot

```python
# Seaborn theme
sns.set_theme(style="whitegrid", context="talk", palette="Set2")

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle("Diamonds Dataset EDA", fontsize=20, weight='bold', y=1.02)

# 1ï¸âƒ£ Scatter Plot: Carat vs Price
sns.scatterplot(
    data=df,
    x="carat",
    y="price",
    hue="cut",
    alpha=0.5,
    edgecolor="k",
    ax=axes[0, 0]
)
axes[0, 0].set_title("Diamond Prices vs Carat")
axes[0, 0].set_xlabel("Carat Weight")
axes[0, 0].set_ylabel("Price ($)")
axes[0, 0].grid(True, linestyle="--", alpha=0.6)

# 2ï¸âƒ£ Histogram: Price Distribution
sns.histplot(
    data=df,
    x="price",
    bins=40,
    kde=True,
    edgecolor="black",
    ax=axes[0, 1],
    color="skyblue"
)
axes[0, 1].set_title("Price Distribution")
axes[0, 1].set_xlabel("Price ($)")
axes[0, 1].set_ylabel("Frequency")
axes[0, 1].grid(True, linestyle="--", alpha=0.6)

# 3ï¸âƒ£ Bar Plot: Average Price by Cut
avg_price_cut = df.groupby("cut")["price"].mean().reset_index()
sns.barplot(
    x="cut",
    y="price",
    data=avg_price_cut,
    edgecolor="black",
    alpha=0.85,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Average Price by Cut")
axes[1, 0].set_xlabel("Cut Type")
axes[1, 0].set_ylabel("Average Price ($)")
axes[1, 0].grid(True, axis="y", linestyle="--", alpha=0.6)

# 4ï¸âƒ£ Box Plot: Price by Cut
sns.boxplot(
    data=df,
    x="cut",
    y="price",
    ax=axes[1, 1],
    palette="Set2",
    showfliers=True,
    linewidth=1.5
)
axes[1, 1].set_title("Price Distribution by Cut")
axes[1, 1].set_xlabel("Cut Type")
axes[1, 1].set_ylabel("Price ($)")
sns.despine(ax=axes[1, 1], top=True, right=True)
axes[1, 1].grid(True, axis="y", linestyle="--", alpha=0.6)

# Adjust layout
plt.tight_layout()
plt.show()
```

## ðŸŒ¸: Annotations(Top 5 Most Expensive Diamonds)

```python
# -----------------------------
# Seaborn scatter plot
# -----------------------------
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x="carat",
    y="price",
    hue="cut",
    alpha=0.5,
    edgecolor="k",
    palette="Set2"
)

# Title and labels
plt.title("Diamond Prices vs Carat (Annotated)", fontsize=16, weight="bold")
plt.xlabel("Carat Weight", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)

# -----------------------------
# Annotations: top 5 most expensive diamonds
# -----------------------------
top5 = df.nlargest(5, "price")
for i, row in top5.iterrows():
    plt.text(
        row["carat"] + 0.02,     # shift a little on x-axis
        row["price"],             # y-axis
        f"${int(row['price']):,}",# formatted price
        fontsize=10,
        color="black",
        weight="bold"
    )

plt.legend(title="Cut Type")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# 1ï¸âƒ£ Built-in Seaborn palettes
# -----------------------------
# Sequential / Standard
deep_palette = sns.color_palette("deep")
muted_palette = sns.color_palette("muted")
bright_palette = sns.color_palette("bright")
pastel_palette = sns.color_palette("pastel")
dark_palette = sns.color_palette("dark")
colorblind_palette = sns.color_palette("colorblind")

# -----------------------------
# 2ï¸âƒ£ Tab palettes (good for categorical data)
# -----------------------------
tab10_palette = sns.color_palette("tab10")
tab20_palette = sns.color_palette("tab20")
tab20b_palette = sns.color_palette("tab20b")
tab20c_palette = sns.color_palette("tab20c")  # additional option

# -----------------------------
# 3ï¸âƒ£ Setting a global palette
# -----------------------------
# Example: set the pastel palette globally
sns.set_palette("pastel")  

# -----------------------------
# 4ï¸âƒ£ Optional: Display palettes
# -----------------------------
def show_palettes(palettes, title="Seaborn Palettes"):
    plt.figure(figsize=(12, 2))
    for i, color in enumerate(palettes):
        plt.bar(i, 1, color=color)
    plt.title(title, fontsize=14)
    plt.axis('off')
    plt.show()

# Show all major palettes
for name, palette in zip(
    ["Deep", "Muted", "Bright", "Pastel", "Dark", "Colorblind", "Tab10", "Tab20", "Tab20b", "Tab20c"],
    [deep_palette, muted_palette, bright_palette, pastel_palette, dark_palette, colorblind_palette, tab10_palette, tab20_palette, tab20b_palette, tab20c_palette]
):
    show_palettes(palette, title=f"{name} Palette")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# List of most popular Seaborn palettes
# -----------------------------
popular_palettes = [
    "deep",
    "muted",
    "bright",
    "pastel",
    "dark",
    "colorblind",
    "tab10",
    "tab20",
    "tab20b",
    "tab20c"
]

# -----------------------------
# Function to display palette forward and reversed
# -----------------------------
def show_palette(palette_name):
    # Forward
    forward = sns.color_palette(palette_name)
    # Reversed
    reverse = forward[::-1]
    
    plt.figure(figsize=(10, 2))
    
    # Forward
    for i, color in enumerate(forward):
        plt.bar(i, 1, color=color)
    plt.title(f"{palette_name} (Forward)", fontsize=14)
    plt.axis('off')
    plt.show()
    
    # Reversed
    plt.figure(figsize=(10, 2))
    for i, color in enumerate(reverse):
        plt.bar(i, 1, color=color)
    plt.title(f"{palette_name} (Reversed)", fontsize=14)
    plt.axis('off')
    plt.show()

# -----------------------------
# Display all popular palettes
# -----------------------------
for palette in popular_palettes:
    show_palette(palette)

# -----------------------------
# Optional: Set global palette
# -----------------------------
# Example: set 'muted' reversed globally
sns.set_palette(sns.color_palette("muted")[::-1])
```
