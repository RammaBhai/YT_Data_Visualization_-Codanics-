```python
# Import Libraries
import plotly.express as px 
import plotly.graph_objects as go
import plotly.io as pio
import plotly.figure_factory as ff
import plotly.offline as pyo
from plotly.subplots import make_subplots
import plotly.colors as pc

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

# Diamonds dataset (raw CSV link from seaborn-data GitHub)
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)
# Shuffle rows
df = df.sample(frac=1, random_state=10)
df.head()

print(f"Orginal Shape(Before):",{df.shape[0]})
print(f"Orginal Shape(Before):",{df.shape[1]})
print(f"{50*'-'}")
df = df.sample(5000)
print(f"New Shape(After):", {df.shape[0]})
print(f"New Shape(After):",{df.shape[1]})
```

## üêî: Scatter Plot with Marginal Histogram

```python
fig = px.scatter(
    df,
    x="carat",
    y="price",
    marginal_x="histogram",
    marginal_y="histogram",
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="carat",
    y="price",
    marginal_x="histogram",
    marginal_y="histogram",
    trendline="ols",
)

fig.update_layout(
    template="plotly_dark",
    title="Scatter Plot with Marginal Histogram",
    xaxis_title="Carat",
    yaxis_title="Price",
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="carat",
    y="price",
    marginal_x="box",
    marginal_y="violin",
    color="cut",
    trendline="ols",
)

fig.update_layout(
    template="plotly_dark",
    title="Scatter Plot with Marginal Histogram",
    xaxis_title="Carat",
    yaxis_title="Price",
)
fig.show()
```

# üêî: Box + violin + Face

```python
fig = px.scatter(
    df,
    x="carat",
    y="price",
    marginal_x="box",
    marginal_y="violin",
    trendline="ols",
    color="cut",
    facet_col="clarity",
    facet_col_wrap=2,
)

fig.update_layout(
    template="plotly_dark",
    title="Scatter Plot with Marginal Histogram",
    xaxis_title="Carat",
    yaxis_title="Price",
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="carat",
    y="price",
    marginal_x="box",
    marginal_y="violin",
    trendline="ols",
    color="cut",
    facet_col="clarity",
    facet_row="color",
    facet_col_wrap=2,
)

fig.update_layout(
    template="plotly_dark",
    title="Scatter Plot with Marginal Histogram",
    xaxis_title="Carat",
    yaxis_title="Price",
    height=1200,
    width=1000
)

fig.show() 
```
