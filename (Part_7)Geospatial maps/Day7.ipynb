```python
import wbdata
import pandas as pd

# -------------------------
# Countries & Indicators
# -------------------------
countries = ["CN", "JP"]
indicators = {
    "SP.POP.TOTL": "Total_Population",
    "NY.GDP.PCAP.CD": "GDP_Per_Capita"
}

dfs = []

for country in countries:
    print(f"Fetching data for {country}...")
    # Fetch data for country without date filters
    data = wbdata.get_dataframe(indicators, country)
    
    # Reset index
    df = data.reset_index()
    
    # Rename columns
    df.rename(columns={"country":"Country", "date":"Year"}, inplace=True)
    
    # Convert Year to numeric
    df["Year"] = pd.to_datetime(df["Year"], errors='coerce').dt.year
    
    # Filter years 1960-2025
    df = df[(df["Year"] >= 1960) & (df["Year"] <= 2025)]
    
    dfs.append(df)

# Combine countries
df_final = pd.concat(dfs, ignore_index=True)

print(df_final.head())
```

```python
import plotly.express as px
import pandas as pd

# -------------------------
# Prepare your data
# -------------------------
df_anim = df_final.copy()  # Make a copy of your dataset

# Normalize population for bubble sizes (optional)
pop_min = df_anim["Total_Population"].min()
pop_max = df_anim["Total_Population"].max()
df_anim["pop_norm"] = (df_anim["Total_Population"] - pop_min) / (pop_max - pop_min) * 50 + 10  # scaled size 10‚Äì60

# Define selected countries to highlight
selected_countries = ["China", "Japan"]

# Add border width for highlighting
df_anim["border_width"] = df_anim["Country"].apply(lambda x: 4 if x in selected_countries else 0.5)

# -------------------------
# Create animated scatter_geo (bubble map)
# -------------------------
fig = px.scatter_geo(
    df_anim,
    locations="Country",
    locationmode="country names",
    size="pop_norm",                   # Bubble size
    color="Total_Population",          # Bubble color
    hover_name="Country",
    hover_data={"Year": True, "Total_Population": ":,"},
    animation_frame="Year",
    projection="natural earth",
    title="Population of Selected Countries Over Time",
    color_continuous_scale=px.colors.sequential.Plasma,
    range_color=[df_anim["Total_Population"].min(), df_anim["Total_Population"].max()],
    size_max=50
)

# -------------------------
# Update traces: bold borders for selected countries
# -------------------------
fig.update_traces(
    marker_line_width=df_anim["border_width"],
    marker_line_color="black"
)

# -------------------------
# Update map appearance
# -------------------------
fig.update_geos(
    visible=True,
    showcoastlines=True,
    showland=True,
    landcolor="lightgray",
    showocean=True,
    oceancolor="LightBlue",
    projection_type="natural earth",
    coastlinecolor="black"
)

# -------------------------
# Add annotations for selected countries (latest year)
# -------------------------
annotations = []
latest_year = df_anim["Year"].max()
for country in selected_countries:
    country_data = df_anim[(df_anim["Country"] == country) & (df_anim["Year"] == latest_year)]
    if not country_data.empty:
        pop_val = int(country_data["Total_Population"].values[0])
        annotations.append(
            dict(
                x=country,
                y=pop_val,
                xref="x",
                yref="y",
                text=f"{country}: {pop_val:,}",
                showarrow=True,
                arrowhead=2,
                arrowsize=1,
                arrowcolor="red",
                font=dict(color="red", size=12)
            )
        )
fig.update_layout(annotations=annotations)

# -------------------------
# Layout adjustments
# -------------------------
fig.update_layout(
    title_font_size=22,
    coloraxis_colorbar=dict(title="Population"),
    margin={"r":0, "t":50, "l":0, "b":0},
    transition={"duration": 500}
)

# -------------------------
# Optional: clean hover template
# -------------------------
fig.update_traces(
    hovertemplate="<b>%{hovertext}</b><br>Year: %{customdata[0]}<br>Population: %{customdata[1]:,}<extra></extra>"
)

# -------------------------
# Show the final animated map
# -------------------------
fig.show()
```

```python
# ==========================================
# üåç Advanced Population Change Over Time
# Highlighting Selected Countries
# ==========================================

import plotly.express as px
import pandas as pd
import numpy as np

# -------------------------
# Settings
# -------------------------
selected_countries = ["China", "Japan"]
bubble_bg_scale = 18
bubble_hi_scale = 45

df_anim = df_final.copy()

# -------------------------
# Data safety & sorting
# -------------------------
df_anim = df_anim.dropna(subset=["Total_Population"])
df_anim = df_anim.sort_values(["Country", "Year"])

# -------------------------
# Log-based population normalization
# -------------------------
df_anim["pop_log"] = np.log10(df_anim["Total_Population"])

pop_min = df_anim["pop_log"].min()
pop_max = df_anim["pop_log"].max()

df_anim["pop_norm"] = (
    (df_anim["pop_log"] - pop_min) / (pop_max - pop_min)
).clip(0.05, 1)

# -------------------------
# Pretty population labels
# -------------------------
df_anim["Pop_Label"] = df_anim["Total_Population"].apply(
    lambda x: f"{x/1e9:.2f}B" if x >= 1e9 else f"{x/1e6:.1f}M"
)

# -------------------------
# Split background & highlighted countries
# -------------------------
df_bg = df_anim[~df_anim["Country"].isin(selected_countries)]
df_hi = df_anim[df_anim["Country"].isin(selected_countries)]

# -------------------------
# Background layer
# -------------------------
fig = px.scatter_geo(
    df_bg,
    locations="Country",
    locationmode="country names",
    size=df_bg["pop_norm"] * bubble_bg_scale + 2,
    color=df_bg["pop_norm"],
    color_continuous_scale="Blues",
    opacity=0.35,
    animation_frame="Year",
    projection="natural earth",
    hover_name="Country",
    hover_data={
        "Year": True,
        "Total_Population": False,
        "Pop_Label": True,
        "pop_norm": False
    },
    title="üåç Global Population Change Over Time"
)

# -------------------------
# Highlight layer
# -------------------------
fig_hi = px.scatter_geo(
    df_hi,
    locations="Country",
    locationmode="country names",
    size=df_hi["pop_norm"] * bubble_hi_scale + 10,
    animation_frame="Year",
    hover_name="Country",
    hover_data={
        "Year": True,
        "Total_Population": False,
        "Pop_Label": True
    }
)

# -------------------------
# Add highlight traces
# -------------------------
for trace in fig_hi.data:
    trace.marker.color = "crimson"
    trace.marker.opacity = 0.95
    trace.marker.line.width = 3.5
    trace.marker.line.color = "black"
    fig.add_trace(trace)

# -------------------------
# Map styling
# -------------------------
fig.update_geos(
    showland=True,
    landcolor="rgb(240,240,240)",
    showocean=True,
    oceancolor="rgb(200,220,255)",
    showcountries=True,
    countrycolor="gray",
    coastlinecolor="black"
)

# -------------------------
# Layout polish
# -------------------------
fig.update_layout(
    title=dict(
        text="üåç Global Population Change Over Time<br>"
             "<sup>China & Japan highlighted for demographic contrast</sup>",
        x=0.5
    ),
    title_font_size=22,
    margin=dict(l=0, r=0, t=70, b=0),
    showlegend=False,
    transition=dict(duration=450, easing="cubic-in-out"),
    coloraxis_showscale=False
)

fig.show()
```

```python
# ==========================================
# üöÄ Plotly Animation ‚Üí GIF (CORRECT WAY)
# ==========================================

import plotly.express as px
import pandas as pd
import imageio
import os
import shutil

# ----------------------------
# üìÇ Paths
# ----------------------------
OUTPUT_DIR = "./07_plotly"
FRAMES_DIR = f"{OUTPUT_DIR}/frames"

os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(FRAMES_DIR, exist_ok=True)

html_file = f"{OUTPUT_DIR}/diamonds_animation.html"
gif_file  = f"{OUTPUT_DIR}/diamonds_animation.gif"

# ----------------------------
# üìä Example Data (World Bank style / demo)
# ----------------------------
df = px.data.gapminder().query("year >= 2000")

# ----------------------------
# üéû Animated Figure
# ----------------------------
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    animation_frame="year",
    animation_group="country",
    log_x=True,
    size_max=55,
    range_x=[100, 100000],
    range_y=[40, 85],
    template="plotly_dark",
    title="üåç Global Development Over Time"
)

# ----------------------------
# üíæ Save Interactive HTML
# ----------------------------
fig.write_html(html_file, include_plotlyjs="cdn")
print(f"‚úÖ Interactive HTML saved at: {html_file}")

# ----------------------------
# üñº Export Frames (PNG)
# ----------------------------
frame_files = []

for i, frame in enumerate(fig.frames):
    fig.update(data=frame.data)
    frame_path = f"{FRAMES_DIR}/frame_{i:03d}.png"

    fig.write_image(
        frame_path,
        format="png",
        width=900,
        height=600,
        scale=2
    )

    frame_files.append(frame_path)

print(f"üñº {len(frame_files)} frames exported")

# ----------------------------
# üé¨ Combine Frames ‚Üí GIF
# ----------------------------
with imageio.get_writer(gif_file, mode="I", duration=0.6) as writer:
    for frame in frame_files:
        writer.append_data(imageio.imread(frame))

print(f"üéâ Animated GIF created: {gif_file}")

# ----------------------------
# üßπ Cleanup (Optional)
# ----------------------------
shutil.rmtree(FRAMES_DIR)
print("üßπ Temporary frames cleaned up")
```
